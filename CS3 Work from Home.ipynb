```python
#create new category for unknown values in race, payer_code, and medical_speciality
data["race"] = data["race"].fillna("UNKNOWN")
data["payer_code"] = data["payer_code"].fillna("UNKNOWN")
data["medical_specialty"] = data["medical_specialty"].fillna("UNKNOWN")
```

```python
##one-hot encoding for variables admission_type_id, discharge_disposition_id, admission_source_id

category_var = ["admission_type_id", "discharge_disposition_id", "admission_source_id"]
data[category_var] = data[category_var].astype("str")
```

```python
cols_num = ['time_in_hospital','num_lab_procedures', 'num_procedures', 'num_medications',
       'number_outpatient', 'number_emergency', 'number_inpatient','number_diagnoses']
cols_cat = ['race', 'gender', 
       'max_glu_serum', 'A1Cresult',
       'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide',
       'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide',
       'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
       'tolazamide', 'insulin',
       'glyburide-metformin', 'glipizide-metformin',
       'glimepiride-pioglitazone', 'metformin-rosiglitazone',
       'metformin-pioglitazone', 'change', 'diabetesMed','payer_code']

cols_cat_num = ['admission_type_id', 'discharge_disposition_id', 'admission_source_id']

data[cols_cat_num] = data[cols_cat_num].astype('str')
```

```python
data_new =  pd.get_dummies(data[cols_cat + cols_cat_num + ['medical_specialty']],drop_first = True)
```

```python
data = pd.concat([data,data_new],axis=1)
```

```python
data.age = data.age.replace({'[0-10)':5, 
          '[10-20)':15, 
          '[20-30)':25, 
          '[30-40)':35, 
          '[40-50)':45, 
          '[50-60)':55,
          '[60-70)':65, 
          '[70-80)':75, 
          '[80-90)':85, 
          '[90-100)':95})
```

```python
#or maybe drop weight na lang
data['has_weight'] = data.weight.notnull().astype('int')
```

## what to with repeating patient numbers??

```python
## removing raw eright column
cols_all_cat = np.array(data.columns)
cols_all_cat = np.delete(cols_all_cat,5)
```

```python
cols_extra = ['age','has_weight']
cols = cols_num + list(cols_all_cat) + cols_extra
```

### Create filtered column dataset

```python
final = data[cols + ['Readmitted?']]
```

```python
final.gender.value_counts()
```

```python
final = final[final.gender!="Unknown/Invalid"]
```

```python
y = final["Readmitted?"]
x = final.drop(["Readmitted?","readmitted","weight",'encounter_id', 'patient_nbr'],axis=1)
```

```python
x.columns
```

```python
x.head()
```

## Logistic Regression

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
logit = LogisticRegression(fit_intercept=True, penalty='l1')
logit.fit(X_train, y_train)
```

```python
logit_pred = logit.predict(X_test)
pd.crosstab(pd.Series(y_test, name = 'Actual'), pd.Series(logit_pred, name = 'Predict'), margins = True)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
print("Accuracy is {0:.2f}".format(accuracy_score(y_test, logit_pred)))
print("Precision is {0:.2f}".format(precision_score(y_test, logit_pred)))
print("Recall is {0:.2f}".format(recall_score(y_test, logit_pred)))
```

Typically, it is better to balance the data in some way to give the positives more weight. There are 3 strategies that are typically utilized
sub-sample the more dominant class:
- use a random subset of the negatives
- over-sample the imbalanced class: use the same positive samples multiple times
- create synthetic positive data
